```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt

optloc = pd.read_csv('./PredictedData/OptimalLocation/PunterOptLoc2020.csv')
rv = pd.read_csv('./PredictedData/RealLocationValues.csv')
```

```python
rv['Field_Val'] = rv['x_land'] - rv['R_avg']*rv['R_len'] + rv['NF_avg']*(rv['p_tb']*(90-rv['x_land'])+rv['ex_ntb'])
rv.loc[(rv['y_land']<0)|(rv['y_land']>160/3), 'Field_Val'] = rv.loc[(rv['y_land']<0)|(rv['y_land']>160/3), 'x_land']
rv.loc[rv['x_land']>110,'Field_Val'] = 90
```

```python
comparison = rv[['gameId','playId','x_land','y_land','Field_Val','R_len','p_tb']].merge(optloc[['gameId','playId','x_land','y_land','Field_Val', 'R_len', 'p_tb','kickerId']], on=['gameId','playId'], suffixes=['','_opt'])
comparison['dFV'] = comparison['Field_Val'] - comparison['Field_Val_opt']
comparison
```

```python
name_dict = pd.read_csv('./data/players.csv')[['nflId','displayName']].set_index('nflId').to_dict()['displayName']
AvgNetYard = comparison.groupby('kickerId')['dFV'].sum().div(comparison.groupby('kickerId').count()['gameId']).sort_values(ascending=False).reset_index()
AvgNetYard['Name'] = AvgNetYard['kickerId'].map(name_dict)
AvgNetYard
```
